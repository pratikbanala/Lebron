```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
df = pd.read_csv("/Users/n/Downloads/Cleaned Project Data")
```

```python
df.rename({"Unnamed: 0":"DATE"},axis=1,inplace=True)
```

```python
df.head()
```

```python
df.index = df["DATE"]
df.drop("DATE",axis=1,inplace=True)
df.index = pd.to_datetime(df.index)
```

```python
df.head()
```

```python
X = df[["DEFRTG","DREB%","MINUTES PLAYED"]]
y = df.PTS
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
df_model = DecisionTreeClassifier()
df_model.fit(X_train,y_train)
df_model.score(X_test,y_test)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
normed = scaler.fit_transform(df).copy()
df_normed = pd.DataFrame(data=normed,columns=df.columns)
```

```python
df_normed.head()
```

```python
X = df_normed[["DEFRTG","DREB%","MINUTES PLAYED"]]
y = df_normed.PTS
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
df_model_norm = DecisionTreeClassifier()
df_model_norm.fit(X_train,y_train)
df_model_norm.score(X_test,y_test)
```

```python
reg_models = [linear_model.HuberRegressor(),
 linear_model.PoissonRegressor(),
 linear_model.RANSACRegressor(),
 linear_model.TweedieRegressor(),
 linear_model.SGDRegressor(),
 linear_model.BayesianRidge(),
 linear_model.LassoLars(),
 linear_model.ARDRegression(),
 linear_model.PassiveAggressiveRegressor(),
 linear_model.TheilSenRegressor(max_subpopulation=10000),
 linear_model.LinearRegression()]
```
